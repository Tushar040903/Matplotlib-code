```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "cricketBat": ["SS","GM","kookabura","SG"],
    "price": [1000,2000,15000,500],
    "Grams": [1100,1000,800,2000]}
```

```python
dataframe = pd.DataFrame(data)
```

```python
plt.xlabel("Price")  # for adding label to x-axis
plt.ylabel("Grams")  # for adding label to y-axis
plt.grid()           # for adding grid to graph
plt.plot(dataframe["price"] , dataframe["Grams"])  # for plotting the graph
```

```python
plt.title("bats")  # to put tittle 
```
