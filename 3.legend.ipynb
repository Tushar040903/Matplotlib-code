```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(5)
b = [2,4,6,8,10]
c = [5,6,7,8,9]
```

```python
fig = plt.figure()
ax = plt.subplot()
```

```python
fig = plt.figure()
ax = plt.subplot()
ax.plot(a , b, 'k--', label = 'frequency')
ax.plot(a , c, 'k:', label = 'periods')
legend = ax.legend(loc = "upper center", fontsize = 'xx-large') # use lower right , lower left , lower center etc
legend.get_frame().set_facecolor('blue') # to change the bgcolor of legend
```
